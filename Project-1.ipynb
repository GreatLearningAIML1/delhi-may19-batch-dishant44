<h1>
4) Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
</h1>
<h2>
Goal_diff_count = GoalsFor - GoalsAgainst
    </h2>

```python
np.dtype(laliga['GoalsFor'])
```

```python
np.dtype(laliga['GoalsAgainst'])
```

```python
laliga[['GoalsFor','GoalsAgainst']] = laliga[['GoalsFor','GoalsAgainst']].apply(pd.to_numeric) 
print(laliga[['GoalsFor','GoalsAgainst']].dtypes)
# to convert dtypes from object to int
```

```python
def Goal_diff_count():
    difference = abs(laliga['GoalsFor'] - laliga['GoalsAgainst']) ## abs is for getting absolute values as here we get some negative values as difference
    laliga['Goal Difference']=difference
    x=laliga[['Team','Goal Difference']]
    print('The teams and their respecti e goal differences are:','\n',x)
    max_goal=laliga['Goal Difference'].max()
    maximum_goal=laliga[laliga["Goal Difference"]==max_goal]
    print('The team which has maximum goal difference is: ' , '\n',maximum_goal[['Team','Goal Difference']])
    min_goal=laliga['Goal Difference'].min()
    minimum_goal=laliga[laliga["Goal Difference"]==min_goal]
    print('The team which has minimum goal difference is: ' , '\n',minimum_goal[['Team','Goal Difference']])
```

```python
Goal_diff_count()
```
